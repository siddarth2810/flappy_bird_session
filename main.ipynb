```python
import pygame
import random

pygame.init()

#setting up width and height of the screen
SCREEN = pygame.display.set_mode((500,750))

BACKGROUND_IMAGE = pygame.image.load('background.jpg')

#bird image
BIRD_IMAGE = pygame.image.load('bird1.png')

bird_x = 75
bird_y = 300
bird_y_change = 0


#PIPES
PIPE_WIDTH = 70
PIPE_HEIGHT = random.randint(150,450) #generate between 150 and 450
PIPE_COLOR  = ("lightgreen")
PIPE_X_CHANGE = -0.2 #Pipe moves towards the left
PIPE_X = 500 #starting at the end of the screen
PIPE_GAP = 180  # Constant gap between top and bottom pipes

def display_pipe(height):
    pygame.draw.rect(SCREEN, PIPE_COLOR, (PIPE_X, 0, PIPE_WIDTH, height))
    bottom_height = 635 - height -PIPE_GAP
    pygame.draw.rect(SCREEN, PIPE_COLOR, (PIPE_X, 635-bottom_height, PIPE_WIDTH, bottom_height))
    
    
def collision_detection(PIPE_X, PIPE_HEIGHT, bird_y, bottom_height):
    if PIPE_X >=75 and PIPE_X <=(75+64):
        if bird_y <= PIPE_HEIGHT or bird_y >= (bottom_height-64):
            return True
    return False
    

while True:
    # Process player inputs.
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            raise SystemExit
            
        #for the functionality of the bird
        if event.type == pygame.KEYDOWN:
            if event.key ==pygame.K_SPACE:
                bird_y_change = -0.6
                
        #for the functionality of the bird
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_SPACE:
                bird_y_change = +0.3
                
             

    SCREEN.fill("orange") 
    
    #Draw background image from 0,0
    SCREEN.blit(BACKGROUND_IMAGE, (0,0))
    
    display_pipe(PIPE_HEIGHT) 
    
    
    collision = collision_detection(PIPE_X, PIPE_HEIGHT, bird_y, PIPE_HEIGHT + PIPE_GAP)
    
    if collision:
        pygame.quit()
        
    
    
    
    bird_y = bird_y + bird_y_change
    
    if bird_y <=0:
        bird_y =0
    if bird_y >=571:
        bird_y = 571
        
    PIPE_X = PIPE_X + PIPE_X_CHANGE
    
    if PIPE_X <=0:
        PIPE_X = 500
        PIPE_HEIGHT = random.randint(200,400)
    
    
    #Draw bird
    SCREEN.blit(BIRD_IMAGE, (bird_x,bird_y))
#     print(bird_y)
    
    pygame.display.update()  #update the screen

pygame.quit()    
```
